# C) Standardise the data matrix and use R to verify that the sample mean vector of the standardised data is composed of zeros and the sample covariance matrix is equal to the sample correlation matrix of the original data.

```r
standardized_airpollution = scale(airpollution)
```

```r
colMeans(standardized_airpollution)
```

```r
apply(standardized_airpollution, 2, var)
```

- We have here that the variance of the transformed/standardized variables is  equal to one (1).

```r
# The covariance matrix of standardized data
var(standardized_airpollution)
```

```r
# The Correlation matrix of the original data
cor(airpollution)
```

Here we can see that the sample covariance matrix of the standardized data is equal to the sample correlation matrix of the original data.

From above we have that several of the original variables have considerably larger means and variances than the other variables which can result in them  dominating principal component analysis based on the raw data.
As such it is more appropriate to base the PCA on the standardised data which is centred with all the variables on a common scale.

B) Using your choice from 2(a), perform a PCA. Write down the first two principal components and give an interpretation.

```r
(pca1 = prcomp(x = airpollution, scale = TRUE))
```

```r
# First PC
pca1$rotation[, 1]
```

From principal component 1 we have:
- The coefficients for all the variables are positive
- The coefficients for SMEAN is the highest
- The coefficient of SMAX and PMEAN are vary similar <br>
This PC may be interpreted as average measure of affect of mean biweekly sulphate reading and mean biweekly suspended particulate reading airpollution

```r
# Seconnd PC
pca1$rotation[, 2]
```

From principal component 2 we have
- The coefficients for all the pollutant measures, aside from SMIN, is negative.
- The coefficient of GE65 is highest, followed by PERWH and NONPOOR. <br>
This PC can be interpreted as average measure of population with white population, population living above poverty line and population above the age of 65.

C) How many principal components would you recommend using in this analysis. Justify your answer.

```r
# Importance of Principal Components
summary(pca1)
```

```r
# Scree Plot
plot(pca1, type="lines", main="")
title(xlab="Component number")
```
